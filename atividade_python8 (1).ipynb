```python
#F)

sql = """SELECT
Funcionario.Primeiro_nome,
Departamento.nome_der,
Funcionario.Fucao 
FROM Funcionario
JOIN Departamento ON Funcionario.cod_Depar = Departamento.codigo ORDER BY Funcionario.Primeiro_nome """
cursor.execute(sql)
cursor.fetchall()
```
